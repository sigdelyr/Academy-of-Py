```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python

# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

```python
# Total schools
total_schools = len(school_data_complete["School ID"].unique())
total_schools
```

```python
# Total students
total_students = len(school_data_complete["Student ID"].unique())
total_students
```

```python
# Total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
# The average math score
average_math_score = student_data["math_score"].mean()
average_math_score
```

```python
# The average reading score
average_reading_score = student_data["reading_score"].mean()
average_reading_score
```

```python
# The overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overal_passing_score = (average_math_score + average_reading_score)/2
overal_passing_score
```

```python
# % of students with a passing math score (70 or greater)
student_data["passing_math"] = student_data["math_score"] >= 70
percent_passing_math = ((student_data["passing_math"]).mean())*100
percent_passing_math
```

```python
# % of students with a passing reading score (70 or greater)
student_data["passing_reading"] = student_data["reading_score"] >=70
percent_passing_reading = ((student_data["passing_reading"]).mean())*100
percent_passing_reading
```

```python
# % Overall Passing (Average of the above two)
overall_passing_rate = (percent_passing_math + percent_passing_reading)/2
print(overall_passing_rate)
```

```python
# Create a dataframe to hold the above results

district_results = [{"Total Schools": total_schools, 
            "Total Students": total_students, 
            "Total Budget": total_budget, 
            "Average Math Score": average_math_score, 
            "Average Reading Score": average_reading_score, 
           "% Passing Math": percent_passing_math,
           "% Passing Reading": percent_passing_reading,
            "% Overall Passing Rate": overall_passing_rate}]

district_summary_table = pd.DataFrame(district_results)

district_summary_table



```

* Create an overview table that summarizes key metrics about each school, including:

```python
# School name
school_group = school_data_complete.groupby("school_name")
school_group
```

```python
# School type
school_type = school_group["type"].unique()
school_type
```

```python
# Total students per school
school_size = school_group["size"].mean()
school_size
```

```python
# Total school budget
school_budget = school_group["budget"].mean()
school_budget
```

```python
# Per student budget
per_student_budget = school_budget / school_size
per_student_budget
```

```python
#Average Math Score per school
school_ave_math_score = (school_group["math_score"].sum()) / school_size
school_ave_math_score




```

```python
#Average reading Score per school
school_ave_reading_score = (school_group["reading_score"].sum()) / school_size
school_ave_reading_score
```

```python
# passing math
math_per_school =  school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name").count()["student_name"]
math_per_school

```

```python
# % passing math
school_percent_pass_math = (math_per_school / school_size) * 100
school_percent_pass_math
```

```python
# passing reading
reading_per_school =  school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name").count()["student_name"]
reading_per_school
```

```python
# % passing reading
school_percent_pass_reading = (reading_per_school / school_size) * 100 
school_percent_pass_reading
```

```python
# Overal passing rate
school_overall_passing_rate = (school_percent_pass_math + school_percent_pass_reading) / 2
school_overall_passing_rate
```

```python
# Let's create a dictionary by using data above to make a dataframe 
school_summary_dict = {"School type": school_type, "Total students": school_size,
                      "Total school budget": school_budget, "Per student budget": per_student_budget,
                      "Average math score": school_ave_math_score, "Average reading score": school_ave_reading_score,
                      "% passing math": school_percent_pass_math, "% passing reading": school_percent_pass_reading,
                      "Overall passing rate": school_overall_passing_rate}


school_summary_df = pd.DataFrame(school_summary_dict)
school_summary_df


```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
school_summary_copy = school_summary_df.copy()
school_summary_copy["Per student budget"] = school_summary_copy["Per student budget"].astype(float)
school_summary_copy["Spending Ranges (Per Student)"] = pd.cut(school_summary_copy["Per student budget"], spending_bins, labels=group_names)

math_per_spending = school_summary_copy.groupby("Spending Ranges (Per Student)").mean()["Average math score"]
reading_per_spending = school_summary_copy.groupby("Spending Ranges (Per Student)").mean()["Average reading score"]
math_percent_spending = school_summary_copy.groupby("Spending Ranges (Per Student)").mean()["% passing math"]
reading_percent_spending = school_summary_copy.groupby("Spending Ranges (Per Student)").mean()["% passing reading"]
overal_passing_spending = school_summary_copy.groupby("Spending Ranges (Per Student)").mean()["Overall passing rate"]

scores_school_spending = {"Average math score": math_per_spending, "Average reading score": reading_per_spending,
                     "% passing math": math_percent_spending, "% passing reading": reading_percent_spending,
                     "Overall passing rate": overal_passing_spending}

scores_school_spending_df = pd.DataFrame(scores_school_spending)
scores_school_spending_df


```
